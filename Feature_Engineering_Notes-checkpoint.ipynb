### Fonksiyonlaştırma

```python
def outlier_thresholds(dataframe, col_name, q1=0.25, q3=0.75):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit
```

```python
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
```

Kategorik değişkenleri seçme ihtiyacımız olacak ileride veya kategorik görünümlü sayısal değişkenleri yada sayısal görünümlü kategorik değişkenleri seçme ihtiyacımız olacak. Bu fonksiyonu özetle ifade etmek gerekirse, ilk önce kategorik değişkenler seçilmiş daha sonra numerik ama kategorik olanlar seçilmiş. Çalışmanın başında cat_th belirlenmiş. Bu categoric threshold demek. 10'dan az bir sınıfa sahipse bir değişken sayısal olsa bile bu değişken bizim için kategorik değişkendir. Yani bir değişkenin içerisindeki sınıf sayısı 10'dan az ise sayısal olsa dahi bu bir kategorik değişken olabilir.
Sonrasında cardinal threshold belirliyoruz. Eğer bir kategorik değişkenin 20'den fazla sınıfı varsa ve tipide kategorikse bu kategorik görünüp kardinaldir. 

cat_cols listesini baştn oluşturuyoruz. Burada numerik görünümlü kategorik değişkenler vardı. Bunları da ekliyoruz. cat_cols'un içerisinde kardinalitesi yüksek değişkenler de vardı, kategorik ama kardinal olmayan değişkenleri seçip onları çıkarıyoruz.

num_cols numerik kolonları ifade etmektedir. Burada tipi object'ten farklı olanları yani int, float olanları num_cols'un içine gelecek. numerik olup kategorik olanları da çıkaracağız.

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
    """
    
    Veri setindeki kategorik, numerik ve kategorik fakat kardinal değişkenlerin isimlerini verir.
    Not: Kategorik değişkenlerin içerisinde numerik görünümlü kategorik değişkenler de dahildir.
    
    Parameters
    ----------
        dataframe: dataframe
                Değişkenlerin isimleri alınmak istenilen veriseti
        cat_th: int, optional
                numerik fakat kategorik olan değişkenler için sınıf eşik değeri
        car_th: int, optional
                kategorik fakat kardinal değişkenler için sınıf eşik değeri
                
    Returns
    -------
        cat_cols: list
                Kategorik değişken listesi
        num_cols: list
                Numerik değişken listesi
        cat_but_car: list
                Kategorik görünümlü kardinal değişken listesi
                
    Examples
    --------
        import seaborn as sns
        df = sns.load_dataset("iris")
        print(grab_col_names(df))
        
    Notes
    -----
        cat_cols + num_cols + cat_but_car = toplam değişken sayısı
        num_but_cat cat_cols'un içerisinde.
        Return olan 3 liste toplamı toplam değişken sayısına eşittir: cat_cols + num_cols + cat_but_car = değişken sayısı
    
    
    """
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes !="O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f'cat_cols: {len(cat_cols)}')
    print(f'num_cols: {len(num_cols)}')
    print(f'cat_but_car: {len(cat_but_car)}')
    print(f'num_but_cat: {len(num_but_cat)}')
    
    return cat_cols, num_cols, cat_but_car
```

### Aykırı Değerlere Erişmek

Daha önce aykırı değerlerin index'lerine erişmiştik. Amacımız bu işlemi fonksiyonlaştırmak. Aykırı değerleri yakalamak bir iş ise yakaladığımızı bildiğimiz bir noktada onların ne olduğuna gidip bir bakma ihtiyacımız olduğunda diğer gözlemler açısından değerlendirmek ise diğer bir iştir. Tekrar tekrar bir quantile yazmak yerine grab_quantile adında bir fonksiyon yazacağız. Eğer bir değişkendeki aykırı değer sayısı 10'dan büyükse yani 10'dan fazla sayıda aykırı değer varsa sadece bize bir ilk 5 gözlemini göstersin, değilse bütün gözlemi göstersin.

```python
def grab_outliers(dataframe, col_name, index=False):
    low, up = outlier_thresholds(dataframe, col_name)
    if dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].shape[0] > 10:
        print(dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].head())
    else:
        print(dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))])
        
    if index:
        outlier_index = dataframe[((dataframe[col_name] < low) | (dataframe[col_name] > up))].index
        return outlier_index
```

Yukarıdaki fonksiyonu tanımlarsak, index bilgisini ön tanımlı olarak False yaptık çünkü index bilgisini şimdilik istemiyoruz. grab_outliers bizim için ilk olarak analiz fonksiyonu.
```python
def remove_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    df_without_outliers = dataframe[~((dataframe[col_name] < low_limit) | (dataframe[col_name] > up_limit))]
    return df_without_outliers
```

### Çok Değişkenli Aykırı Değer Analizi (Local Outlier Factor)

Elimizde iki tane değişken olduğunu varsayalım. Değişkenlerden bir tanesi yaş değişkeni diğeri ise evlilik sayısı değişkeni olsun. Evlilik sayısı değişkenine odaklanacak olursak 3 kere evlenmiş olmak aykırı bir değer olmayabilir. Tabiki sayı yükseldikçe aykırı değer ihtimali ortaya çıkacaktır. Yaş değişkeni için bakacak olursak 17 yaş yaş değişkeni için anormal bir sayı değil gibi duruyor. 17 yaşında olup 3 defa evlenmiş olmak anormal bir durum mudur? Yani tek başına aykırı olamayacak bazı değerler birlikte ele alındığında bu durum aykırılık yaratıyor olabilir. Buna çok değişkenli etki denir. 

LOF yöntemi gözlemleri bulundukları konumda yoğunluk tabanlı skorlayarak buna göre aykırı değer tanımı yapabilmemizi sağlar. Bir noktanın lokal yoğunluğu demek ilgili noktanın etrafındaki komşuluklar demektir. Bir nokta komşularının yoğunluğundan anlamlı bir şekilde düşük ise bu durumda bu nokta daha seyrek bir bölgededir ve aykırı bir değerdir yorumu yapılabilir.

LOF yöntemi bize bir skor verir ve bu skor 1'e ne kadar yakın olursa o kadar iyidir der. 1'den uzaklaşan ilgili gözlemin outlier olabileceği ile ilgili fikir verir.

Sadece eksik değere sahip değişkenlerin isimlerini seçmenin yolu list compherension ile yapılmaktadır.

```python
def missing_values_table(dataframe, na_name=False):
    na_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]
    n_miss = dataframe[na_columns].isnull().sum().sort_values(ascending=False)
    ratio = (dataframe[na_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False)
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis = 1, keys = ['n_miss', 'ratio'])
    print(missing_df, end="\n")
    if na_name:
        return na_columns
```

Eksik veriye ilk bakış için "missing_values_table" adında bir fonksiyon tanımlıyoruz. Bu fonksiyonun tanımlı argümanında "dataframe", ikinci argümanında ise "na_columns" adında eksik değerlerin isimlerinin bulunduğu değişkenleri verir. İlk satırda ilgili veri seti tanımlandığında bu veri seti içerisindeki eksik değere sahip değişkenlerinin isimlerinin seçildiği bir argüman var. Sonrasında eksik değer sayısını bulmak için "n_miss" adında bir argüman oluşturuyoruz. "n_miss" argümanının altında "ratio" adında eksik değerlerle ilgili oran veren bir argüman tanımlıyoruz. En son olarak bunları "missing_df" adında bir veri setine çevirip "concat" işlemi ile birleştiriyoruz.

## Encoding Scaling

Encode ifadesi değişkenlerin temsil şekillerinin değiştirilmesi demektir. Label Encoding ise bir kategorik değişkenin sınıfları label'lerdir ve buradaki labellerı yeniden encode etmek demektir. Bir veri seti içerisindeki değişkenin gözlemleri nominal kategorik değişken ise bunu label encoding uygulayamayız. Böyle bir değişkeni one hot encoding işlemine sokabiliriz. 

### Label Encoding & Binary Encoding

Kategorik değişkenin iki sınıfı varsa ve bu sınıflar 1-0 olarak kodlanırsa buna binary encoding denir. İkiden fazla sınıfı varsa label encoding denir.

LabelEncoder algoritmasını çağırıp le adında bir değişkene atıyoruz. Sonrasında veri setimiz içerisindeki "Sex" değişkenini le'ye transform et diyerek veri seti içerisindeki kategorik değişkeni encode ediyoruz.

Bu dönüştürme işlemlerinde 1-0 verdiğimiz değerlerin hangisi olduğunu unuttuğumuzda "inverse_transform" metodunu kullanarak değerlerimizi görebiliyoruz.

```python
def label_encoder(dataframe, binary_col):
    labelencoder = LabelEncoder()
    dataframe[binary_col] = labelencoder.fit_transform(dataframe[binary_col])
    return dataframe
```

Burada rare kategoriler ile bağımlı değişken arasındaki ilişkinin analizini yapıyor olacağız. "NAME_INCOME_TYPE" değişkenini "TARGET" değişkenine göre groupby ortalamasını alıyoruz ve gözlemliyoruz. Bussinessman sınıfının "TARGET" değişkeni açısından ortalamasının 0 olduğunu görüyoruz. Burada 1 ve 1'e yakın olan rakamlar kredi ödeyememeyi ifade etmektedir. Bussinessman sınıfının kredi ödeyebilme durumlarında herhangi bir problem yaşamadığına yakın gibi görünüyor. Student sınıfının da kredi ödeyebilitesinde problem olmadığına yakın olarak gözlemliyoruz. Unemployment sınıfının ödeyememe durumuna daha yakın olduğu gözlemleniyor.

```python
def rare_analyser(dataframe, target, cat_cols):
    for col in cat_cols:
        print(col, ":", len(dataframe[col].value_counts()))
        print(pd.DataFrame({"COUNT": dataframe[col].value_counts(),
                           "RATIO": dataframe[col].value_counts() / len(dataframe),
                           "TARGET_MEAN": dataframe.groupby(col)[target].mean()}), end="\n\n\n")
```

Bu işlemleri bir araya getirecek bir fonksiyon yazdık. Bu fonksiyona bağımlı değişkeni ve kategorik değişkenleri gönderdiğimizde yukarıda daha önce yapmış olduğumuz işlemleri getirecek bir fonksiyon oldu. İlk olarak ilgili kategorik değişkenin kaç sınıf olduğu ile ilgili bilgiyi, ikincisi sınıf frekansları, üçüncüsü sınıf frekans oranları ve dördüncü olarak da bağımlı değişkene göre groupby işlemi yaptık.

```python
rare_analyser(df, "TARGET", cat_cols)
```

Elimizdeki bütün kategorik değişkenler için rare analinizi gerçekleştirmiş olduk.

```python
def rare_encoder(dataframe, rare_perc):
    temp_df = dataframe.copy()
    rare_columns = [col for col in temp_df.columns if temp_df[col].dtype == "O" and (temp_df[col].value_counts() / len(temp_df) < rare_perc).any(axis=None)]
    for var in rare_columns:
        tmp = temp_df[var].value_counts() / len(temp_df)
        rare_labels = tmp[tmp < rare_perc].index
        temp_df[var] = np.where(temp_df[var].isin(rare_labels), "Rare", temp_df[var])
    return temp_df
```

```python
new_df = rare_encoder(df, 0.01)
```

```python
rare_analyser(new_df, "TARGET", cat_cols)
```

Burada 0.01'in altındaki bütün sınıflar bir araya getirdik ve bu sınıflara rare sınıflar dedik. "rare_encoder" adında bir fonksiyon oluşturduk, bu dataframe'in içerisinde değişiklik yapılacağı için temp_df adında bir isim oluşturarak buraya kopyaladık. Fonksiyonun başında "rare_perc" adında bir argüman tanımladık ve rare_columns'un içerisinde "rare_perc" argümanına girilen değerden daha düşük sayıda herhangi bir kategorik değişkenin sınıf oranı varsa ve aynı zamanda bu bir kategorik değişken ise bunları "rare_columns" olarak getir dedik. 
Sonra bu rare_columns'larda gezip bunların "value_counts( )"'unu alıyoruz ve toplam gözlem sayısına bölerek "temp_df" veri setinin içerisindeki bir değişken için sınıf oranlarını hesapladık. İlgili değişkeni başta girdiğimiz "rare_perc" argümanının sayısına indirgedikten sonra kalan indexleri "rare_labels" adında bir yer oluşturarak tutuyoruz. Akabinde "temp_df" içerisinde gezdiğimiz "rare_columns"'ların içerisinde herhangi bir "rare_labels" gözlemlersek buralara "Rare" yazdırıyoruz, değilse normal bırakıyoruz.

```python
from statsmodels.stats.proportion import proportions_ztest

test_stat, pvalue = proportions_ztest(count=[df.loc[df["NEW_CABIN_BOOL"] == 1, "Survived"].sum(),
                                            df.loc[df["NEW_CABIN_BOOL"] == 0, "Survived"].sum()],
                                            
                                            nobs=[df.loc[df["NEW_CABIN_BOOL"] == 1, "Survived"].shape[0],
                                                 df.loc[df["NEW_CABIN_BOOL"] == 0, "Survived"].shape[0]])
print('Test Stats = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

Burada iki grubun oranını kıyaslama testi yaptık. Yeni oluşturduğumuz bir feature var ve bu feature'ın istatistiki olarak bağımlı değişken ile arasındaki ilişkiyi merak ettiğimiz için oran testi yapıyoruz. Gözlenme sayısı olarak kabin numarası olanlar kaç kişi, kabin numarası olmayanlar kaç kişi diyerek sorguladık. "Test Stats = 9.4597, p-value = 0.0000" bu çıktıya göre anlamlı bir fark var gibi duruyor.
